```python
def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = plt.contourf(xx, yy, Z, **params)
    return out  




d0, d1 = data[:, 0], data[:, 1]
xx, yy = make_meshgrid(d0, d1)  
  
  
  
plot_contours(clf, xx, yy,cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(d0, d1,c= label, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xticks(())
plt.yticks(())
plt.title("svm")  
  
```

```python
from sklearn import datasets
from sklearn import svm
import pandas as pd
import matplotlib.pyplot as plt


digit_data = datasets.load_digits()

print(digit_data.keys())

#features = pd.dataFrame(digit_data.data,columns = digit_data.)


print("no. data",digit_data.data.shape)
print("images",digit_data.images.shape)

plt.subplot(121)
plt.imshow(digit_data.data[10].reshape(8,8))
plt.subplot(122)
plt.imshow(digit_data.images[10])

#looks like data and images are both same 
print("target",digit_data.target.shape)
print("target_name",digit_data.target_names.shape)


# keyword : target represents the label 1797
# keyword : target_names the label is 10 can be used as column


#train test data
train_data = digit_data.data[0:1700]
train_label = digit_data.target[0:1700]
test_data = digit_data.data[1700:1797]
test_label = digit_data.target[1700:1797]





digit_clf = svm.SVC(gamma = 0.0001 ,C = 100)

digit_clf.fit(train_data,train_label)


print("score ",digit_clf.score(test_data,test_label))




```
